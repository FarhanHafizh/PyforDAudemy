### Assign values to entire columns 

```python
nfl_clipboard['Stadium'] = "Levi's Stadium"
nfl_clipboard
```

```python
nfl_clipboard['Num_Stadium'] = np.arange(10)
nfl_clipboard
```

### What if you want to add a series to a dataframe

```python
Stadiums = pd.Series(["Levi's Stadium",'AT&T Stadium'],index=[9,0])
Stadiums
```

```python
nfl_clipboard['Stadium'] = Stadiums
nfl_clipboard
```

### Deleting column

```python
del nfl_clipboard['Stadium']
nfl_clipboard
```

```python
nfl_clipboard.drop('Num_Stadium',axis=1,inplace=True)
nfl_clipboard
```
